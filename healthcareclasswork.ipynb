```r
# Remove any duplicate entries based on the Name and Date of Admission.
library(psych)
library(dplyr)
data <- read.csv("healthcare_dataset.csv")
# dim(data)
# n_distinct(data)
duplicate <- duplicated(select(data, Name, Date.of.Admission))
duplicate_rows <- data[duplicate, ]
print(duplicate_rows)
data_unique <- distinct(data, Name, Date.of.Admission, .keep_all = TRUE)
write.csv(data_unique, "cleaned_data.csv", row.names = FALSE)
```

```r
# Convert the Date of Admission and Discharge Date to proper date format in R
library(psych)
library(dplyr)
library(lubridate)
data <- read.csv("healthcare_dataset.csv")
# date <- select(data, Date.of.Admission, Discharge.Date)
converted_date <- mutate(data,
    Admission = as.Date(Date.of.Admission, "%Y-%m-%d"),
    Discharge = as.Date(Discharge.Date, "%Y-%m-%d")
  )
converted_date
```

```r
# Calculate the length of stay for each patient (difference between Discharge Date and Date of Admission)
library(psych)
library(dplyr)
library(lubridate)
data <- read.csv("healthcare_dataset.csv")
data <-
  mutate(data,
    Date.of.Admission = as.Date(Date.of.Admission, format = "%Y-%m-%d"),
    Discharge.Date = as.Date(Discharge.Date, format = "%Y-%m-%d")
  )
stay <- mutate(data, Length.of.Stay = as.numeric(Discharge.Date - Date.of.Admission))
select (stay, Name, Length.of.Stay)
write.csv(stay, "stay_length_added_dataset.csv")
```

```r
# What is the average length of stay?
library(psych)
library(dplyr)
data <- read.csv("stay_length_added_dataset.csv")
summary(data)
mean(data$Length.of.Stay)
```

```r
# Which medical condition has the longest average length of stay
library(psych)
library(dplyr)
data <- read.csv("stay_length_added_dataset.csv")
df <- group_by(data, Medical.Condition)
summarise(df, avg_stay = mean(Length.of.Stay, na.rm = TRUE))
```
